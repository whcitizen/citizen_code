```python
#データの読込
import numpy as np
import pandas as pd
import codecs

with codecs.open("X_train.csv","r","Shift-JIS","ignore") as file:
    df = pd.read_csv(file)

df
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

#dfをarrayに変換
activity = df.as_matrix()

#最適なクラスタ数の計算
distortions = []

for i  in range(1,10):               
    km = KMeans(n_clusters=i).fit(activity) 
    distortions.append(km.inertia_)   

#クラスタ内誤差平方和のグラフ表示
plt.plot(range(1,10),distortions,marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')

#画像として保存
plt.savefig("activity_k.png")
```

```python
with codecs.open("y_train.csv","r","Shift-JIS","ignore") as file:
    category = pd.read_csv(file)

category = category.as_matrix()

category
```

```python
#PCAで次元削減
pca = PCA(n_components=2)
activity_r = pca.fit_transform(activity)

# 結果を散布図にプロット
plt.figure()
for i in range(0, len(category)):
    if category[i] == 1:
        plt.scatter(activity_r[i, 0], activity_r[i, 1], c='red')
    elif category[i] == 2:
        plt.scatter(activity_r[i, 0], activity_r[i, 1], c='green')
    elif category[i] == 3:
        plt.scatter(activity_r[i, 0], activity_r[i, 1], c='Cyan')
    elif category[i] == 4:
        plt.scatter(activity_r[i, 0], activity_r[i, 1], c='orange')
    elif category[i] == 5:
        plt.scatter(activity_r[i, 0], activity_r[i, 1], c='pink')
    elif category[i] == 6:
        plt.scatter(activity_r[i, 0], activity_r[i, 1], c='purple')

plt.title("activity pca")
plt.xlim([-8,15])
plt.ylim([-6,6])
        
#画像として保存
plt.savefig("activity_pca_output.png")
```
